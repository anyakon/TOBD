```python
import xlsxwriter
import xlwings as xw
import numpy as np
```

1. На листе "Рецептура" файла `себестоимостьА_в1.xlsx` для области "Пшеничный хлеб" рассчитать себестоимость всех видов продукции.

```python
x = xw.Book("/Users\Admin\Downloads\себестоимостьА_в1 (1).xlsx")
x
```

```python
sht = xw.sheets['Рецептура']
sht
```

```python
x = sht.range('G7:O10').options(np.array).value
c = sht.range('G14:O14').options(np.array).value
```

```python
seb = np.nan_to_num(x*c).sum(axis=1)
seb
```

2. Результаты расчетов 1.1 сохранить в отдельном столбце области "Пшеничный хлеб"

```python
sht.range('T7:T10').options(transpose = True).value = seb
```

```python
sht.range('T6').value = 'Себестоимость'
```
