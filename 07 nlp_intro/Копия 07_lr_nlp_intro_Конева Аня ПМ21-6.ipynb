```python
spisok=[]
pattern = re.compile(r'[\d]')#удалим цифры из описаний

for descriptions in f1['preprocessed_descriptions']:
    descriptions_new=re.sub(pattern,' ',str(descriptions))
    words = nltk.word_tokenize(str(descriptions_new).lower())
    spisok.append(words)
spisok
```

1.3 Напишите функцию, которая для заданного слова `word` возвращает `k` ближайших к нему слов из списка `words` (близость слов измеряется с помощью расстояния Левенштейна)

```python
def k_bliz(word,words,k):
  dist=[]
  for wor in words:
    data_words = {'word': wor, f'distance_to_{word}': edit_distance(word,wor)}
    dist.append(data_words)
  newlist = sorted(dist, key=lambda k: k['distance_to_telling']) 
  bliz=[]
  for i in range(k):
    bliz.append(newlist[i])
  return bliz
    
```

```python
k_bliz(word='telling',words=unique,k=10)
```
